```python
# importing libraries
 
import cv2
import matplotlib.pyplot as plt
```

```python
# import the images
org_meme = cv2.imread('./img/manpc.jpg')

# convert to RGB
rgb_meme = cv2.cvtColor(org_meme, cv2.COLOR_BGR2RGB)

# show with matplotlib
plt.figure(figsize=(20,20))
plt.imshow(rgb_meme)
```

```python
# create a copy of the image to work on
meme_copy = rgb_meme.copy()

# define coordinates and color for the rectangle enclosing the monitor
bot_corner_m = (700,400)
top_corner_m = (400,60)
red = (255,0,0)

# draw the rectangle
cv2.rectangle(meme_copy, bot_corner_m, top_corner_m, red, 2)

# write the text
cv2.putText(meme_copy, 'Monitor', (400, 420), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,0), 2, cv2.LINE_AA)

# define coordinates and color for the rectangle enclosing the face
bot_corner_f = (200,280)
top_corner_f = (50,50)
red = (255,0,0)

# draw the rectangle
cv2.rectangle(meme_copy, bot_corner_f, top_corner_f, red, 2)

# write the text
cv2.putText(meme_copy, 'Face', (50, 300), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,0), 2, cv2.LINE_AA)

# define coordinates and color for another rectangle
bot_corner_d = (350,100)
top_corner_d = (250,50)
red = (255,0,0)

# draw the rectangle
cv2.rectangle(meme_copy, bot_corner_d, top_corner_d, red, 2)

# write the text
cv2.putText(meme_copy, 'Demon', (250, 115), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,0), 2, cv2.LINE_AA)

# define coordinates and color for the rectangle for the punchline
bot_corner_t = (800,550)
top_corner_t = (0, 490)

# draw the rectangle
cv2.rectangle(meme_copy, bot_corner_t, top_corner_t, (255,255,255), -1)

# write the text
cv2.putText(meme_copy, 'Paranormal CV', (250,515), cv2.FONT_HERSHEY_TRIPLEX, 1, (0,0,0), 2, cv2.LINE_AA)

# show with matplotlib
plt.figure(figsize=(15,15))
plt.imshow(meme_copy)
```
