```python
# importing libraries

import cv2
import matplotlib.pyplot as plt
```

```python
def sketching(imgpath):
    # read the image from the path given as an argument
    img = cv2.imread(imgpath)
    
    # apply a GaussianBlur filter
    img_blur = cv2.GaussianBlur(img, (5,5), 0)
    
    # Apply Canny to detect edges and line
    sketch = cv2.Canny(img_blur, 100, 200)
    
    # Use a bitwise_not since Canny returns black background and white lines
    sketch_not = cv2.bitwise_not(sketch)
    
    # show image with matplotlib
    plt.figure(figsize=(10,10))
    plt.imshow(sketch_not, cmap='gray')
```

```python
sketching('./img/landscape.jpg')
```
