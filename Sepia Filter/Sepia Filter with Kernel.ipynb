```python
# import libraries

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# helper function to show images in matplotlib, since OpenCV loads in BGR, this invert the channels

def imgshow(img):
    plt.figure(figsize=(10,10))
    plt.imshow((img[:,:,::-1]))
```

```python
def sepia_kernel(imgpath):
    # read the image from the path given as an argument
    img = cv2.imread(imgpath)
    
    # define the kernel for the transformation
    kernel = np.array([ [0.272, 0.534, 0.131],
                     [0.349, 0.686, 0.168],
                     [0.393, 0.769, 0.189] ])
    
    # apply the kernel to the image
    sepia = cv2.transform(img, kernel)
    
    # show the image
    imgshow(sepia)
```

```python
cat = cv2.imread('./img/cat.jpg')
imgshow(cat)
```
