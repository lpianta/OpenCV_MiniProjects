```python
# import libraries

import cv2
import matplotlib.pyplot as plt
```

```python
# helper function to show images in matplotlib, since OpenCV loads in BGR, this invert the channels

def imgshow(img):
    plt.figure(figsize=(10,10))
    plt.imshow((img[:,:,::-1]))
```

```python
def sepia_filter(imgpath):
    
    # read the image from the path given in the argument
    img = cv2.imread(imgpath)
    
    # split the image in the three color channels
    b, g, r = cv2.split(img)
    
    # apply the transformation needed at the colors
    tb =  (0.272*r) + (0.534*g) + (0.131*b)
    tg = (0.349*r) + (0.686*g) + (0.168*b)
    tr = (0.393*r) + (0.769*g) + (0.189*b)
    
    # make sure the new value are in the right datatype
    tb, tg, tr = tb.astype('int'), tg.astype('int'), tr.astype('int')
    
    # iterate through the pixel to make sure they don't go over the threshold
    for i in range(len(tb)):
        for j in range(len(tb)):
            if tb[i][j] > 255:
                tb[i][j] = 255
    for i in range(len(tb)):
            for j in range(len(tb)):
                if tb[i][j] > 255:
                    tb[i][j] = 255
    for i in range(len(tb)):
            for j in range(len(tb)):
                if tb[i][j] > 255:
                    tb[i][j] = 255
    
    # merge the three modified channel
    sepia_img = cv2.merge([tb, tg, tr])
    
    # show the image
    imgshow(sepia_img)
```
