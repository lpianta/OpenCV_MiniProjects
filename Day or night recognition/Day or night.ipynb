```python
# import libraries

import cv2
import matplotlib.pyplot as plt
```

```python
# helper function to show images in matplotlib, since OpenCV loads in BGR, this invert the channels

def imgshow(img):
    plt.figure(figsize=(10,10))
    plt.imshow((img[:,:,::-1]))
```

```python
def day_night(imgpath):
    # read the image in the imagepath we pass as an argument of the function
    img = cv2.imread(imgpath)
    
    # create a copy of the image to work with
    img_copy = img.copy()
    
    # convert color spaces to hsv
    img_hsv = cv2.cvtColor(img_copy, cv2.COLOR_BGR2HSV)
    
    #split the channel
    h, s, v = cv2.split(img_hsv)
    
    # check the Value mean according to the threshold
    if v.mean() > 100:
        imgshow(img)
        print('Day time')
    else:
        imgshow(img)
        print('Night time')
        
```
