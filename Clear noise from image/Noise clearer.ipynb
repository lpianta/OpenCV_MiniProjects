```python
# import libraries

import cv2
import matplotlib.pyplot as plt
```

```python
# helper function to show images in matplotlib, since OpenCV loads in BGR, this invert the channels

def imgshow(img):
    plt.figure(figsize=(10,10))
    plt.imshow((img[:,:,::-1]))
```

```python
def noise_cleaning(imgpath):
    # read the image from the imgpath given as argument
    img = cv2.imread(imgpath)
    
    #apply a little gaussian filter to blur the noise
    gaussian_img = cv2.GaussianBlur(img, (3,3), 0)
    
    #call to the function to show the image
    imgshow(gaussian_img)
```

```python
# import and show the original image

img = cv2.imread('./img/noisy.png')
imgshow(img)
```

```python
noise_cleaning('./img/noisy.png')
```
